#### Using the AND operator to satisfy two conditions at once.
#### If we want to look at places where installs is 100+ and Rating<3

```python
playstore_100=playstore['Installs']=='100+'
playstore_3=playstore['Rating']<3
playstore[playstore_100 & playstore_3]
```

#### Using the OR operator to satisfy two conditions at once.
#### If we want to look at places where installs is 100+ or rating>5.

```python
playstore_100=playstore['Installs']=='100+'
playstore_5=playstore['Rating']<5
playstore[playstore_100 & playstore_5]
```

#### Now we will look how many null values are here

```python
playstore.isnull().sum()
```

#### compensatig for the null values

```python
playstore['Rating']=playstore['Rating'].fillna(playstore['Rating'].mean())
```

```python
playstore['Type']=playstore['Type'].fillna(playstore['Type'].ffill())
```

```python
playstore['Content Rating']=playstore['Content Rating'].fillna(playstore['Content Rating'].ffill())
```

```python
playstore['Current Ver']=playstore['Current Ver'].fillna(playstore['Current Ver'].ffill())
```

```python
playstore['Android Ver']=playstore['Android Ver'].fillna(playstore['Android Ver'].ffill())
```

#### Checking if there are any null values left

```python
playstore.isnull().sum()
```

```python
playstore.info()
```

### Now as we can see there are no null value in our dataset since we have filled them all.

```python
playstore1=playstore[['Category','Installs']]
playstore1
```

```python
avg_category=playstore.groupby('Category')['Rating'].mean()
#it will take the mean of rating for each category
avg_category
```

```python
ax=avg_category.plot.bar(title='avg rating wrt category')
ax.set_ylabel('avg rating')
ax.set_xlabel('category')
```

```python
avg_install=playstore.groupby('Installs')['Rating'].mean()
#it will take the mean of rating for each category of install
avg_install
```

```python
ax=avg_install.plot.bar(title='avg mean rating for number of installs')
ax.set_ylabel('average rating')
ax.set_xlabel('installs')
```

```python
avg_price=playstore.groupby('Content Rating')['Rating'].mean()
avg_price
```

```python
ax=avg_price.plot.bar(title='avg mean rating for different content rating')
ax.set_ylabel('mean rating')
ax.set_ylabel('content rating')
```
